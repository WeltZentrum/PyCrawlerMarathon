## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
url = 'https://www.ettoday.net/news/news-list-2020-7-5-0.htm'
browser.get(url)
```

```python
# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
import datetime

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if date.hour >= 19 and date.hour <= 21:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
url = 'https://www.ettoday.net/news/news-list-2020-7-2-0.htm'
browser.get(url)
```

```python
# 每個兩秒鐘自動往下滑
import time
for i in range(60):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

import datetime

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if date.hour >= 15 and date.hour <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
