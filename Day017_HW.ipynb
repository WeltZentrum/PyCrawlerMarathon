### 1. 動態網頁跟靜態網頁的差別是什麼？原本靜態網頁的做法會產生什麼問題或是缺點？

Ans:靜態網頁是透過每一次使用者請求，後端會產生一次網頁回傳。請求與回傳是一對一形式的。
    動態網頁可以在任何時間點上都可以發送請求給後端，取的後端傳送的資料。

Ans:靜態網頁缺點是無發更新資料，必須得再發送請求，後端才會回傳一次資料。
