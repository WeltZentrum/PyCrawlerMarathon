```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
data = json.loads(r.text)

#一次回傳多少資料
print("總共", len(data), "筆資料")
    
```

```python
print("每一筆資料包含多少欄位")
number = 1


for i in data[0]:
    print(number,i)
    number+=1
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
number = 1
for i in data:
    number +=1
    print(number, "標題", i['title'],'\n貼文時間', i['createdAt'], '\n留言人數', i['commentCount'], '\n按讚人數', i['likeCount'], '\n') 
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''

r_1 = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
data_1 = json.loads(r_1.text)

comment = 0
like = 0
comment_1 =0
like_1 =0

for i in data:
    comment += i['commentCount']
    sum = comment
    like += i['likeCount']
    sum_1 = like

for i in data_1:
    comment_1 += i['commentCount']
    sum_2 = comment_1
    like_1 += i['likeCount']
    sum_3 = like_1
    
print("熱門文章", '\n平均留言人數:', sum / len(data), '\n平均按讚人數:', sum_1 / len(data),'\n')
print("非熱門文章", '\n平均留言人數:', sum_2 / len(data), '\n平均按讚人數:', sum_3 / len(data))
```
