```python
### ① 取出今天所有的發文
```

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

for i in soup.find(class_="part_list_2").find_all('h3'):
    print(i.find(class_="date").text, i.find(class_='tag').text, '\n', i.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
for i in soup.find(class_="part_list_2").find('h3'):
    print(i)
```

```python
news = {}

for i in soup.find(class_="part_list_2").find_all('h3'):
    if i.find(class_='tag').text in news:
        news[i.find(class_='tag').text].append({"時間" : i.find(class_='date').text ,'文章': i.find(target="_blank").text})
    else:
        news[i.find(class_='tag').text] = [{'時間' : i.find(class_='date').text ,'文章': i.find(target="_blank").text}]
```

```python
for i in news:
    print('\n', i,'\n')
    for t in news[i]:
        print(t)
```

### ③ 哪一個類別的文章最多

```python
title = ''
count = 0
for i in news:
    if len(news[i]) > count:
        title = i
        count = len(news[i])
        
print(title, '類別總共有',count, '篇')
```
