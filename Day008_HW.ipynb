### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.zhihu.com/explore'

'''
Your Code Here
'''
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get(url, headers = headers)
r.encoding = 'utf-8'
soup = BeautifulSoup(r.text, "html.parser")

print(soup)
```
