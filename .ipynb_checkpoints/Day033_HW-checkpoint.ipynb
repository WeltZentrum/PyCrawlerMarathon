### 為獲取前存取

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.ptt.cc/bbs/Beauty/M.1564040419.A.F79.html', verify = False)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

### 在請求中帶入Cookie

```python
res = requests.get('https://www.ptt.cc/bbs/Beauty/M.1564040419.A.F79.html',verify = False, cookies={'over18': '1'})
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text)
```
