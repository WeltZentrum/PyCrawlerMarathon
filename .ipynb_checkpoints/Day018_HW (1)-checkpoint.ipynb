## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

```python
### 利用查看開發者工具觀察 Headers

from IPython.display import Image

Image("Headers.png", width=800, height=800)
```

```python
import requests
url = 'http://www.zhihu.com'

headers = {'user-agent': 'my-app/0.0.1'}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```

### 2. http://aicoin.cn

```python
import requests

url = 'http://aicoin.cn/'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

```python
### 利用查看開發者工具觀察 Headers

from IPython.display import Image

Image('Headers2.png', width = 800, height= 800)
```

```python
import requests
url = 'http://aicoin.cn/'

headers = {'user-agent': 'my-app/0.0.1'}


r = requests.get(url, headers=headers)
print(r.text[0:600])
```
