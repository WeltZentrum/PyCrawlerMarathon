```python
#引用函式庫
import requests
import time
import pandas as pd
import numpy as np
import jieba
import jieba.analyse
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from wordcloud import WordCloud
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

# 透過 requests / BeautifulSoup / selenium 進行動態爬蟲

```python
# 進行網路抓取
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.cupoy.com/newsfeed/topstory")
```

```python
# 每個兩秒鐘自動往下滑
number = 0
Number = []
title = []
Articles = []
time_before = []
link = []

#透過 jieba 等斷詞將⽂章拆解
corpus = []

while True:
    
    #
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 新頁面高度
    browser.execute_script("return document.body.scrollHeigh")
    time.sleep(2)
    
    # 取得資料
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    hot_news = soup.find_all('div', class_='sc-eEieub sc-iuDHTM emMJDZ')
    
    for i in hot_news:
        number+=1
        Number.append(number)
        title.append(i.find(class_="sc-gacfCG bPSpUf").text)
        Articles.append(i.find(class_="sc-erNlkL sc-ekulBa gyyrZF").text)
        time_before.append(i.find(class_= "sc-cqPOvA gYtNga").text)
        link.append(i.find('a').get('href'))
        corpus += jieba.cut(i.find(class_="sc-erNlkL sc-ekulBa gyyrZF").text)
        

    print(f"目前加載{number}篇新聞")
    
    if number == 500:
        print("爬蟲結束",'\n',"有500篇新聞")
        break
    
    if number > 500:
        print("爬蟲結束", ':',"超過500篇新聞")
        break
        
browser.close()
```

# 整理成 pandas.DataFrame

```python
df = pd.DataFrame(columns=['Number','title','Articles', 'time_before','link'])
df['Number'] = Number
df['title'] = title
df['Articles'] = Articles
df['time_before'] = time_before
df['link'] = link

#取 500 筆資料
df = df.drop(labels=range(500, 510),axis=0) #axis=0 表示按行删除，删除索引值是第 501 行至第 510 行的正行数据
df.tail()
```

```python
#計算分類次數
df_temp = df.title.value_counts().rename_axis('title').reset_index(name='counts')
df_temp
```
