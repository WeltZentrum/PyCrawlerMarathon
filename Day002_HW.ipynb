### 比較一下範例檔案中的「File I/O」與「CSV Reader」讀出來的內容有什麼差異

```python
'''
Your Code

'''
import csv

with open("./Data/example.csv", 'r',encoding="utf-8") as fh:
    f = fh.read()

with open('./Data/example.csv', newline='', encoding="UTF-8") as csvfile:
    data = csv.reader(csvfile)
    for i in data:
        print(i[:3], f[:52])
```

### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 1. 取出班次一的每一個時間

'''
Your Code
'''     
number = 0
with open('./data/example.csv', newline='', encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    class1 = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for time in class1:
        number+=1
        print(number, time[5])
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

'''
Your Code
'''
data = []
with open('./Data/example.csv', newline='', encoding='UTF-8') as csvfile:
    class1 = csv.reader(csvfile)
    for time in class1:
        data.append(time[5])

print(data[:77], type(data))
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

'''
Your Code
'''
data_1 = []
data_2 = []
data_3 = []
data_4 = []

with open('./Data/example.csv', newline='', encoding='UTF-8') as csvfile:
    class1 = csv.reader(csvfile)
    for time in class1:
        data.append(time[5])
        data_1.append(time[6])
        data_2.append(time[7])
        data_3.append(time[8])
        data_4.append(time[9])
        

print(data[:77], type(data))
print(data_1[:77], type(data_1))
print(data_2[:77], type(data_2))
print(data_3[:77], type(data_3))
print(data_4[:77], type(data_4))
```
